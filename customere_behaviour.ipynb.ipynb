```python
import pandas as pd
```

```python
df= pd.read_csv('customer_shopping.csv')
```

```python
df.head()
```

```python
# this help in getting infomation about coluumns
df.info()
```

```python
 # help getting statistical information of numerical columns
df.describe()
```

```python
# with this we can check total null value in each coloums
df.isnull().sum()
```

```python
# filling null values 
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(" ","_")
df = df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# create column age group
labels=["young_adult","adult","middle_age","senior"]

df['age_group']= pd.qcut(df['age'] , q=4 , labels = labels)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create column purchase frequency days
a1={
    'Fortnightly':14, 'Weekly':7, 'Annually':365, 'Quarterly':90, 'Bi-Weekly':14,
       'Monthly':30, 'Every 3 Months':90}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(a1)
```

```python
df.info()
```

```python
df["purchase_frequency_days"].head(10)
```

```python
# column  and  have same data so we remove one of them

(df["discount_applied"] == df["promo_code_used"]).all()
```

```python
df= df.drop("promo_code_used",axis=1)
```
